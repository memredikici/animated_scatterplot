```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import imageio
```

## Reading Fertility and Life Expectancy tables

```python
fert = pd.read_csv('./data/gapminder_total_fertility.csv', index_col=0)
fert
```

```python
life = pd.read_excel('./data/gapminder_lifeexpectancy.xlsx', index_col=0)
life
```

## Data Wrangling and Cleaning

```python
life.columns
```

```python
#Table of Fertilit's columns are consist of String Objects. It's needed to be transformed to integer to be merged.

fert.columns = fert.columns.astype(int)
```

```python
fert.index.name = 'country'
fert
```

```python
fert = fert.reset_index()
fert
```

```python
#Wide format to Long format

fert = fert.melt(id_vars='country',
                 var_name='year',
                 value_name='fertility_rate'
                )
```

```python
life.index.name = 'country'
life
```

```python
life = life.reset_index()
life
```

```python
life = life.melt(id_vars='country',
                 var_name='year',
                 value_name='life_expectancy'
                )
```

```python
life
```

```python
population = pd.read_excel('./data/gapminder_population.xlsx', index_col=0, usecols="A:CD")
population
```

```python
population.columns = population.columns.astype(int)
```

```python
population = population.reset_index()
population
```

```python
population = population.melt(id_vars='Total population',
                 var_name='year',
                 value_name='population'
                )
```

```python
population
```

```python
# change the column name
population.rename(columns={'Total population':'country'}, inplace=True)
population.head()
```

```python
df = fert.merge(population)
```

```python
df = df.merge(life)
df
```

### Plotting the correlation of Fertility Rate and Life Expectancy between years 1960-2015

```python
#Understanding Annotation only for year 1960

df_year = df.loc[df['year']==1960]
plt.figure(figsize= (10, 5))
plt.axis((0, 10, 4, 90))
plt.xlabel('Fertility Rate')
plt.ylabel('Life Expectancy')
plt.title('Correlation of Life Expectancy - Fertility Rate ', fontsize = 20, loc = 'center')
plt.style.use('seaborn-whitegrid')
plt.annotate('1960',xy=(18,2), fontsize='120', color='grey', alpha=0.3)
g=sns.scatterplot(x='fertility_rate',
                y='life_expectancy',
                data=df_year,
                hue='continent',
                size='population',
                sizes=(20,450)
               )

plt.annotate('1960',xy=(0.5,10), fontsize='70', color='grey', alpha=0.3)

countries = ['Turkey','China','United States']
for country in countries:
    x = df_year[df_year['country']==country]['fertility_rate']
    y = df_year[df_year['country']==country]['life_expectancy']
    x_text = x
    y_text = y + 8
    plt.annotate(country,xy=(x,y),
                 xytext=(x_text,y_text),
                 arrowprops=dict(facecolor='black', shrink=0.02),
                 fontsize='14'
                )

h,l = g.get_legend_handles_labels()
lgd= plt.legend(h[0:7],l[0:13],
                bbox_to_anchor=(1.05, 1),
                loc=2,
                borderaxespad=0.
               )
```

```python
for year in range(1960,2016):
    df_year = df.loc[df['year']==year]
    
    plt.figure(figsize= (10, 5.5))
    plt.axis((0, 10, 4, 90))
    plt.xlabel('Fertility Rate')
    plt.ylabel('Life Expectancy')
    plt.title('Correlation of Life Expectancy - Fertility Rate ', fontsize = 20, loc = 'center')
    plt.style.use('seaborn-whitegrid')
    
    g=sns.scatterplot(x='fertility_rate',
                    y='life_expectancy',
                    data=df_year,
                    hue='continent',
                    size='population',
                   )
    #Annotate the year
    plt.annotate(f'{year}', xy=(0.5,10), fontsize='100', color='grey', alpha=0.3)
    h,l = g.get_legend_handles_labels()
    lgd= plt.legend(h[0:7],l[0:13],
                    bbox_to_anchor=(1.05, 1),
                    loc=2,
                    borderaxespad=0.
                   )
    countries = ['Turkey','China','United States']
    for country in countries:
        x = df_year[df_year['country']==country]['fertility_rate']
        y = df_year[df_year['country']==country]['life_expectancy']
        x_text = x
        y_text = y + 8
        plt.annotate(country,xy=(x,y),
                     xytext=(x_text,y_text),
                     arrowprops=dict(facecolor='black', shrink=0.02),
                     fontsize='14'
                    )
    
    plt.savefig(f'./pictures/{year}.png',
                bbox_extra_artists=(lgd,),
                bbox_inches='tight'
               )
    plt.close()
```

```python
#Creating .gif

images = []

for i in range(1960, 2016):
    filename = f'./pictures/{i}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('animated_scatterplot.gif', images, fps=10)
```
